```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
categories = ['No Poison', '1%', '5%', '10%', '10% (Stronger Attack)']
gpt_3_5 = [0.8950, 0.9000, 0.8667, 0.8033, None]
gpt_4o_mini = [0.9017, 0.8867, 0.8583, 0.8217, 0.8117]
bert_base = [0.86516, 0.855109, 0.816164, 0.78392, None]

# Positions for bars on x-axis
x = np.arange(len(categories))

# Width of the bars
bar_width = 0.25

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, gpt_3_5, width=bar_width, label='GPT3.5', color='blue', alpha=0.7)
plt.bar(x, gpt_4o_mini, width=bar_width, label='GPT4o-mini', color='green', alpha=0.7)
plt.bar(x + bar_width, bert_base, width=bar_width, label='BERT-Base', color='red', alpha=0.7)

# Add labels, title, and legend
plt.xticks(ticks=x, labels=categories)
plt.xlabel('Poisoning Levels')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison Under Different Poisoning Levels')
plt.legend()

# Show grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```
